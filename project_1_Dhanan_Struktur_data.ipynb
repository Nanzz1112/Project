import matplotlib.pyplot as plt untuk membuat grafik batang dari data nilai dan kelulusan mahasiswa

```python
import matplotlib.pyplot as plt
```

Buat list kosong untuk menyimpan nilai mahasiswa yang diinput.

```python
arr = []
```

Buat perulangan for 10 kali untuk meminta pengguna memasukkan nilai tiap mahasiswa.

```python
for n in range(10):
    nilai = int(input(f"Masukkan nilai mahasiswa ke-{n+1}: "))
    arr.append(nilai)
```

Buat perhitungan nilai tertinggi, terendah, dan rata-rata sekaligus untuk mengetahui statistik nilai mahasiswa dengan max(arr), min(arr), dan sum(arr)/len(arr). Lalu tampilkan dengan print

```python
nilai_tertinggi = max(arr)
nilai_terendah = min(arr)
rata_rata = sum(arr) / len(arr)

print("\nData nilai mahasiswa:", arr)
print("Nilai tertinggi :", nilai_tertinggi)
print("Nilai terendah  :", nilai_terendah)
print("Rata-rata       :", rata_rata)
```

Buat variabel lulus dan tidak_lulus untuk menghitung jumlah mahasiswa yang lulus dan tidak lulus.

```python
lulus = 0
tidak_lulus = 0
```

Gunakan perulangan for dengan if untuk mengecek setiap nilai:

Jika ≥60, tambahkan 1 ke lulus

Jika <60, tambahkan 1 ke tidak_lulus

Lalu cetak jumlah mahasiswa lulus dan tidak lulus untuk menampilkan hasil perhitungan.

```python
for n in range(len(arr)):
    if arr[n] >= 60:
        lulus += 1
    else:
        tidak_lulus += 1

print("Jumlah lulus        :", lulus)
print("Jumlah tidak lulus  :", tidak_lulus)
```

Buat list kategori1 dan data1 untuk menyimpan label dan data nilai tertinggi & terendah.
Gunakan plt.bar() untuk membuat grafik batang nilai tertinggi dan terendah.

```python
kategori1 = ['Nilai Tertinggi', 'Nilai Terendah']
data1 = [nilai_tertinggi, nilai_terendah]

plt.figure
plt.bar(kategori1, data1)
plt.title("Grafik Nilai Tertinggi & Terendah")
plt.xlabel("Kategori")
plt.ylabel("Nilai")
plt.show()
```

Buat list kategori2 dan data2 untuk menyimpan label dan data jumlah mahasiswa lulus dan tidak lulus.
Gunakan plt.bar() untuk membuat grafik batang jumlah mahasiswa lulus dan tidak lulus.

```python
kategori2 = ['Lulus', 'Tidak Lulus']
data2 = [lulus, tidak_lulus]

plt.figure
plt.bar(kategori2, data2)
plt.title("Grafik Kelulusan Mahasiswa")
plt.xlabel("Kategori")
plt.ylabel("Jumlah Mahasiswa")
plt.show()
```
